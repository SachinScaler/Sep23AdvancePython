x^3-> find derivative at x 2
```
f(x)=x^3
f'x= 3*x^2
f'2 = 3*(2)^2
=12
```



````python
def ab_initio_method(x1, fx):
    h = 0.000001
    der = (fx(x1+h)-fx(x1))/h
    return der


func = lambda x: x**3 # f_x = x**3
ab_initio_method(2, func)
````

````python
func(2)
````

````python
import math
func = lambda x: math.log(x)*math.sin(x**2)

ab_initio_method(3, func)
````

### plot any function

````python
import numpy as np
import matplotlib.pyplot as plt
def plot_func(func):
    x = np.linspace(-10,10,1000)
    y =[func(i) for i in x]


    deriv = [ab_initio_method(i, func) for i in x]


    plt.plot(x,y)
    plt.plot(x,deriv)

    plt.legend(["function", "derivative"])
    plt.show()

func = lambda x: x**2
plot_func(func)
````
